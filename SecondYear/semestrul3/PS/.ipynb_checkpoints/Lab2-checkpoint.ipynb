```python
from random import randint
#help('random.randint')
#set?
```

```python
#subpunctul a)
N = int(input("Introduceti numarul de persoane: "))

def zile_nastere(nr_persoane):
    list = []
    for i in range(0, nr_persoane) :
        list.append(randint(1,365))
    return list

def zile_duplicate(lista):
    return len(lista) != len(set(lista))
                
def simulare(nr_simulari, nr_persoane):
    cp = 0
    for i in range(0, nr_simulari):
        lista_birthdays = zile_nastere(nr_persoane)
        if (zile_duplicate(lista_birthdays) == True):
            cp = cp +1
    return cp / nr_simulari
            
simulare(1000, N)
```

```python
#subpunctul b)
def probabilitate_zile_nastere(nr_persoane):
    p = 1
    for k in range(nr_persoane):
        p = p * (365 - k)/365
    return 1 - p

probabilitate_zile_nastere(24)
```

```python
from matplotlib.pyplot import plot, grid, title, show 
title('Plot simulari') 
xs = range(2, 51) 
ys = [simulare(1000, x) for x in xs] 
plot(xs,ys,'r*') 
show()
```

```python
title('Plot probabilitate') 
xs = range(2, 51) 
ys = [probabilitate_zile_nastere(x)for x in xs] 
plot(xs,ys,'r*') 
grid() 
show()
```

```python
from matplotlib.pyplot import axis, plot, show, Circle
from random import random 
from math import dist 
#help('random.random') 
#help('math.dist')

def patrat1(puncte):
    axis('square') 
    axis((0, 1, 0, 1)) 
    cnt = 0
    for _ in range (puncte):
        F=[random(),random()] 
        if dist(F, [0.5, 0.5]) < 0.5:
            cnt = cnt + 1
            plot(F[0],F[1],'ro') 
        else:
            plot(F[0],F[1],'bo')
    show()

def patrat2(puncte):
    axis('square') 
    axis((0, 1, 0, 1)) 
    cnt = 0
    A,B,C,D,E = [0,0],[1,0],[1,1],[0,1],[0.5,0.5]
    for _ in range (puncte):
        F=[random(),random()] 
        if dist(F, [0.5, 0.5]) < min(dist(F,A), dist(F,B), dist(F,C), dist(F,D)):
            cnt = cnt + 1
            plot(F[0],F[1],'ro') 
        else:
            plot(F[0],F[1],'bo')
    show()
    
patrat2(5000)
```
